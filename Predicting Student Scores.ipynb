# Importing Libraries

```python
import math
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

# Loading Data

```python
df= pd.read_csv('student_scores.csv')
```

```python
df.head()
```

```python
df.info()
```

# Visualization

```python
plt.scatter(df['Hours'], df['Scores'])
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

# Training the Model

```python
X = df['Hours'].values.reshape(-1, 1)
y = df['Scores'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_pred = model.predict(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': (y_pred)})
print(df)
```

```python
print(f'Train data RMSE: {math.sqrt(mean_squared_error(y_train, X_pred)):.2f}')
print(f'Train data MSE: { mean_squared_error(y_train, X_pred):.2f}')
print(f'Train data MAE: {mean_absolute_error(y_train, X_pred):.2f}')
print("--"* 50)
print(f'Test data RMSE: {math.sqrt(mean_squared_error(y_test, y_pred)):.2f}')
print(f'Test data MSE: {mean_squared_error(y_test, y_pred):.2f}')
print(f'Test data MAE: {mean_absolute_error(y_test, y_pred):.2f}')
```

# Making Predictions

```python
hours = np.array([[9.25]])
predicted_score = model.predict(hours)
print(f'Predicted Score for 9.25 hours/day: {predicted_score[0]:.2f}')
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color='red')
plt.title('Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```
